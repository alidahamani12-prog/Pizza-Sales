## A. The Campus Calculator

```python
#1_Operations performed by the campus calculator
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b == 0:
        return "Error"
    return a / b
```

```python
#2_User's choice of operation
while True:
       print("Campus Calculator")
       print("1. Add")
       print("2. Subtract")
       print("3. Multiply")
       print("4. Divide")
       print("5. Exit")
       
       choice = input("Choose operation (1-5): ")
       
       if choice == '5':
           print("Goodbye")
           break
```

## B. The Class Result Assistant

```python
#This was an alternative way I found would be perfect for the B part of the task2
#I had this idea during the additional time
#I wanted to do this part again using this approach, but I got stucked somewhere

students = {}
num_students = int(input("How many students? "))

for i in range(num_students):
    name = input(f"Enter student {i+1} name: ")
    num_subjects = int(input(f"How many subjects does {name} study: "))
    
    subject_scores = {}

    for j in range(num_subjects):
        subject = input(f"Enter {name} subject {j+1} name: ")
        score = float(input(f"What is the score for {subject}: "))
        subject_scores[subject] = score
    
students[name] = subject_scores
    
```

### The use of conditional statements for grading

```python
  def assign_grade(average):
       if average >= 90:
           return 'A'
       elif average >= 80:
           return 'B'
       elif average >= 70:
           return 'C'
       elif average >= 60:
           return 'D'
       elif average >= 50:
           return 'E'
       else:
           return 'F'
```
